```python
import cv2
import imutils
import pytesseract
from matplotlib import pyplot as plt
```

```python

image = cv2.imread('license4.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = imutils.resize(image, width=300 )
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image,cmap='gray')
```

```python
gray_image = cv2.bilateralFilter(gray_image, 11, 17, 17) 
plt.imshow(gray_image,cmap='gray')
```

```python
#thresh_image1= cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            #cv2.THRESH_BINARY_INV,15,8)
#gray_image = cv2.GaussianBlur(gray_image,(5,5),0)
x,thresh_image1= cv2.threshold(gray_image, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)

#img = cv2.medianBlur(gray_image,5)

plt.imshow(thresh_image1,cmap='gray')
```

```python
#edged = cv2.Canny(thresh_image1,500,500) 
#plt.imshow(edged,cmap='gray')
```

```python
cnts,new = cv2.findContours(thresh_image1.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
image1=image.copy()
cv2.drawContours(image1,cnts,-1,(0,255,0),1)
plt.imshow(image1)
```

```python
cnts1 = sorted(cnts, key = cv2.contourArea, reverse = True) [:30]
screenCnt = None
image2 = image.copy()
cv2.drawContours(image2,cnts1,-1,(0,255,0),1)
plt.imshow(image2)
```

```python
i=7
for c in cnts1:
        perimeter = cv2.arcLength(c, True)
        approx = cv2.approxPolyDP(c, 0.018 * perimeter, True)
        if len(approx) == 4: 
                screenCnt = approx
                x,y,w,h = cv2.boundingRect(c) 
                new_img=image[y:y+h,x:x+w]
                #cv2.imwrite('./'+str(i)+'.jpeg',new_img)
                i+=1
                break
```

```python
screenCnt
```

```python
cv2.drawContours(image, [screenCnt], -1, (0, 255, 0), 1)
plt.imshow(image)
```

```python
plt.title("License plate")
plt.imshow(new_img)
plate = pytesseract.image_to_string(new_img, lang='eng')
print("Number plate is:", plate)
```
